```python
#import meteostat_tempreature as mt       # if you wanna run the weather extraction, uncomment this line of code
import matplotlib.pyplot as plt,geopandas as gpd,shapefile,pandas as pd 
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist

plt.style.use('ggplot')
import numpy as np
import cartopy.crs as ccrs
import cartopy.feature
```

Visualise the local authorities considered to model the aggregated heat demands

```python
#set the file path
fp = 'data/LA_UK/LAD_MAY_2022_UK_BFE_V3.shp'
# Read file using gpd.read_file()
LA_UK = gpd.read_file(fp)
```

```python
LA_UK
```

Cluster the local authorities to 29 regions which can mode the aggregate heat demand for the district heating. For the clustering purpose the KMeans algorithm is implemented and the heat nodes are used as a centroid for the KMeans clustering

```python
# find the coordinates for local authorities and the coordinates of the centroid
#modell='reduced'
modell='zonal'

LA_UK_=[]
for i in range(len(LA_UK)):
    LA_UK_coordinates=LA_UK_.append([LA_UK.iloc[i]['LAT'],LA_UK.iloc[i]['LONG']])
LA_UK_coordinates=np.asarray(LA_UK_)

if modell=='zonal':
    df_network = pd.read_csv('data/buses_ZM.csv')
elif modell=='reduced':
    df_network = pd.read_csv('data/buses.csv')

gb_nodes=[]
for i in range(len(df_network[:29])):
    heat_node_centroids=gb_nodes.append([df_network.iloc[i]['y'],df_network.iloc[i]['x']])
heat_node_centroids=np.array(gb_nodes)
    
```

Remove the Orkeny islands, the northern ireland local authorities

```python
condition = (LA_UK['LAD22NM'] == 'Orkney Islands')  # Specify the condition
LA_UK = LA_UK.drop(LA_UK[condition].index)

condition = (LA_UK['LAD22NM'] == 'Shetland Islands')  # Specify the condition
LA_UK = LA_UK.drop(LA_UK[condition].index)

# exclude the Northern ireland nodes, since we are modelling for GB
for i in range(11):
    condition = (LA_UK['LAD22CD'] == 'N0900000'+str(i+1))  # Specify the condition
    LA_UK = LA_UK.drop(LA_UK[condition].index)
LA_UK = LA_UK.drop(LA_UK[(LA_UK['LAD22CD'] == 'N09000010')].index)
LA_UK = LA_UK.drop(LA_UK[(LA_UK['LAD22CD'] == 'N09000011')].index)
# exclude the isle isalnds
LA_UK = LA_UK.drop(LA_UK[(LA_UK['LAD22NM'] == 'Na h-Eileanan Siar')].index)
```

Using KMeans clustering for the Geopands dataframe based on the geometry points

```python
# Converting the geometry column in a GeoPandas DataFrame from polygons to geographic latitude and longitude coordinates (points),using the centroid property of the polygons
LA_UK_points = LA_UK.copy()

# Convert the 'geometry' column to centroid points
LA_UK_points['geometry'] = LA_UK_points['geometry'].centroid

# Convert the geometry column to longitude and latitude coordinates
LA_UK_points = LA_UK_points.set_geometry('geometry')
LA_UK_points = LA_UK_points.to_crs('EPSG:4326')
```

```python
LA_UK_points
```

Apply the KMeans Clustering to find the clusters in the local authroity

```python
points = LA_UK_points['geometry'].apply(lambda geom: (geom.y, geom.x))
coordinates = np.array(list(points))
centroids=heat_node_centroids
kmeans = KMeans(n_clusters=len(centroids), init=centroids, n_init=1)
clusters = kmeans.fit_predict(coordinates)

LA_UK_points['cluster'] = clusters

LA_UK_points.plot(column='cluster', categorical=True, legend=True, figsize=(15, 59))
plt.legend(bbox_to_anchor=(1.0, 1.0))
plt.show()
```

```python
# Define a color map for the clusters
num_clusters = len(np.unique(clusters))
color_map = plt.get_cmap('tab20', num_clusters)
# Create convex hulls for each cluster
convex_hulls = LA_UK_points.groupby('cluster')['geometry'].apply(lambda x: x.unary_union.convex_hull)

# Plot the shaded convex hulls and scatter points for each cluster
fig, ax = plt.subplots(figsize=(30, 30))
ax = fig.add_subplot(1, 1, 1,
                     projection=ccrs.PlateCarree())

extent = [-8.09782, 2.40511, 60, 49.5]
ax.set_extent(extent)
ax.add_feature(cartopy.feature.OCEAN)
ax.add_feature(cartopy.feature.LAND)
for cluster_label, convex_hull in convex_hulls.items():
    cluster_points = LA_UK_points[LA_UK_points['cluster'] == cluster_label]
    cluster_points.plot(ax=ax, color=color_map(cluster_label), label=f'Cluster {cluster_label}', markersize=10)
    gpd.GeoSeries(convex_hull).plot(ax=ax, facecolor=color_map(cluster_label), alpha=1, linewidth=0)
plt.legend(bbox_to_anchor=(1.0, 1.0))
plt.show()
```
